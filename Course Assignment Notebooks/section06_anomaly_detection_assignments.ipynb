## 1. Isolation Forests

```python
# 1a. Open the tripadvisor_reviews.csv file
import pandas as pd

df = pd.read_csv('../Data/tripadvisor_reviews.csv')
df.head()
```

```python
# 1b. Remove the user_id column
# df.drop(columns=['user_id'])
X = df.iloc[:,1:]
X.head()
```

```python
# 1c. View the min and max of each rating
X.describe()
```

```python
# 2. Visualize the data using a seaborn pair plot
import seaborn as sns

sns.pairplot(
    X,
    height=3
);
```

```python
# 3. Fit an Isolation Forest model using a contamination of 1%
from sklearn.ensemble import IsolationForest

model = IsolationForest(contamination=0.01)
```

```python
model.fit(X)
```

```python
df['anomaly'] = model.predict(X)
df.head()
```

```python
# 4. Visualize the anomalies on the seaborn pair plot
sns.pairplot(
    df,
    hue='anomaly',
    palette='bright',
    height=3
);
```

```python
# 5. Where do you notice anomalies in the pair plot?
# nightlife people hate our town for activities
```

```python
# 6. Modify the contamination to 0.5% to capture fewer anomalies and visualize the differences
model2 = IsolationForest(contamination=0.05)
model2.fit(X)
```

```python
df['anomaly2'] = model2.predict(X)
df.head()
```

```python
sns.pairplot(
    df.drop(columns=['anomaly']),
    hue='anomaly',
    palette='bright',
    height=3
);
```
